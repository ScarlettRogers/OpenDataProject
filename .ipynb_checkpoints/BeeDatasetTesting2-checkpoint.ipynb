```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
import seaborn as sns
import os
import shapely
import fiona

%matplotlib inline
plt.rcParams['figure.figsize'] = (30, 20)
```

```python
loss_data = pd.read_excel("colony_loss.xlsx")
loss_data.sample(5)
```

```python
def PercentagesToFloats(num):
    return "{}%".format(num*100)

loss_data.columns = [x.replace('\xa0', '') for x in loss_data.columns] # replaces \xa0 with nothing
loss_data.columns = [x.replace(' ', '') for x in loss_data.columns]# replaces space with nothing

loss_data.columns = [x.replace('State', 'NAME') for x in loss_data.columns]# changes column name to match with later geojson
loss_data.dropna(how='any',axis='index', inplace=True)# removes all rows with a NaN value
loss_data = loss_data.set_index("NAME")
loss_data = loss_data.drop('MultiStateOperation', axis=0)
# allows for column referencing using df
# e.g. df.State     df.TotalAnnualLoss
# loss_data.rename(columns={ loss_data.columns[-1]: "Index" }, inplace = True)
loss_data.sample(50)
```

```python
def new_range(old_min, old_max, old_value):
#     minimum_value = newplot['TotalAnnualLoss'].min()
#     maximum_value = newplot['TotalAnnualLoss'].max()
    new_min = 0
    new_max = 1000
    new_value = ( (old_value - old_min) / (old_max - old_min) ) * (new_max - new_min) + new_min
```
